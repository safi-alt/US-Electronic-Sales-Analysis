## Q1. what was the best month for sales and how much was earned that month?

### Before we answer this question we need to augment the data so the question can be answered properly

## Clean up the data ðŸ§½ 

```python
# Checking NAN values
nan_df = all_data[all_data.isna().any(axis=1)]
nan_df.head()
```

```python
## Dropped NAN values
all_data = all_data.dropna(how='all')
```

```python
## Finding 'Or' and deleting it
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
all_data
```

```python
# add month column
all_data['month'] = all_data['Order Date'].str[0:2]
all_data['month'] = all_data['month'].astype("int32")
all_data.head()
```

# Add a Sales Column

```python
## Convert column to the corect type
all_data["Quantity Ordered"] = pd.to_numeric(all_data["Quantity Ordered"])
all_data['Price Each'] =  pd.to_numeric(all_data["Price Each"])
```

```python
all_data.head()
```

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
```

```python
all_data.head()
```

# Best month for sales and what is the sales amount

```python
results = all_data.groupby('month').sum()
results
```

```python
import matplotlib.pyplot as plt

months = range(1,13)

plt.bar(months, results['Sales'])
plt.xticks(months)
plt.xlabel('Month number')
plt.ylabel('Sales in USD ($)')
plt.show()
```

# Which city has thw highest sales and how much was the sales

```python
# Add a city column
all_data['City'] = all_data['Purchase Address'].apply(lambda x: x.split(',')[1])

all_data.head()
```

```python
# alternative way of extracting city

def get_city(address):
    return address.split(',')[1]

def get_state(state):
    return state.split(',')[2].split(' ')[1]


all_data['City'] = all_data['Purchase Address'].apply(lambda x: get_city(x) + ' (' + get_state(x) + ')')

all_data.head()
```

```python
results = all_data.groupby('City').sum()
results
```

```python
import matplotlib.pyplot as plt

cities = [city for city, df in all_data.groupby("City")]

plt.bar(cities, results['Sales'])
plt.xticks(cities, rotation = 'vertical', size=8)
plt.yticks(rotation = 'vertical', size=8)
plt.xlabel('City Name')
plt.ylabel('Sales in USD ($)')
plt.show()
```

# What produts are often most sold together

```python
df = all_data[all_data['Order ID'].duplicated(keep=False)]
```

```python
df.head(20)
```

```python
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
```

```python
df.head()
```

```python
df = df[['Order ID','Grouped']].drop_duplicates()
```

```python
df.head(20)
```

```python
from itertools import combinations
from collections import Counter
```

```python
count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))

for key, value in count.most_common(10):
    print(key,value)
```

# What product sold the most and why do you think it sold the most

```python
product_group = all_data.groupby('Product')

quantity_ordered = product_group.sum()['Quantity Ordered']
```

```python
products = [product for product, df in product_group]

plt.bar(products, quantity_ordered)
plt.ylabel("Quantity ordered")
plt.xlabel("Products")
plt.xticks(products, rotation='vertical', size=8)
plt.show()
```

```python
prices = all_data.groupby("Product").mean()['Price Each']
```

```python
print(prices)
```

```python
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products,quantity_ordered, color='g')
ax2.plot(products, prices, 'b-')

ax1.set_xlabel("Product Name")
ax1.set_ylabel("Qauntity Ordered", color='g')
ax2.set_ylabel("Price ($)", color='b')
ax1.set_xticklabels(products, rotation='vertical', size=8)

plt.show()
```
